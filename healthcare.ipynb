```python
                                                                  ###HEALTHCARE###
# Description:To live a healthy life for sunita who is 65+ years of age.the person have to go through various health tests
# and various blood tests.the data which we obtain from the blood test are used to analyse the important aspects like blood pressure,
# sugar level hemoglobin levels,kidney function,cholestrol level,thryroid level.Here i have used the sample data to analyse the result
# for various tests sunita have to undergo and this tests will help here to stay healthy.here i have used various tests and they are:

#1. BMI :here require data is weight and height
#2 heartrate:here require data is age
#3 blood pressure:for blood presuure the systolic and distolic are used to anaylse
#4 cholestrol:for this we used the LDL and  HDL which are low protein lipoprotiens and high protein lipoprotiens
#5 uric acid:uric acid levels
#6 thyroid:for this i used TSH which is throyoid simulating hormone
#7 diabetic:this  test  includes the HbA1c test and and this is used to test whether patient is diabetic or not
#8 kidney-for this i have used eGFR which is estimated glomerular filtration rate
#9 hemoglobin level 


# process data for subject1
subject1 = ['SUNITA', 50, 1.62, 66, 120, 80, 90, 75, 10, 1.4, 35,100, 12]
initials, weight, height, age, systolic, diastolic,LDL,HDL,uric_acid ,TSH, HbA1c,eGFR,hemoglobin = subject1 

# Calculate BMI
bmi = int(weight / height**2)
# Caculate predicted maximum heart rate
max_HR = 208 - 0.7 * age


# Caculate blood pressure risk 
if systolic >= 120 and systolic < 130 and diastolic < 80:
    bprisk = 'elevated BP'
elif (systolic >= 130 and systolic < 140) or (diastolic >= 80 and diastolic < 90):
    bprisk = 'stage 1 hypertension'
elif systolic >= 140 or diastolic >= 90:
    bprisk = 'stage 2 hypertension'
else:
    bprisk = 'invalid values'


#for cholestrol
if LDL<200 and HDL>60:
  cholestrol = "sunita you have a healthy heart"
elif 100<LDL<159 and 50<HDL<59:
  cholestrol = "sunita your cholestrol levels are high"
elif LDL>160 and HDL<50:
  cholestrol = "sunita your cholestrol levels are too high"
else:
  cholestrol = "invalid"


#for uric acid 
if uric_acid<1.5:
   uric_acid = " sunita you have a low uric acid level"
elif 1.5<uric_acid<60:
  uric_acid = " sunita your uric acid level is normal"
elif uric_acid>7.0:
  uric_acid = " sunita your uric acid levels are high"
else:
  uric_acid="invalid"


#for thyroid check
if 0.5<TSH<5.0:
     thyroid = " sunita your thyroid levels are normal"
else:
     thyroid = " sunita your thyroid levels are high "


#for diabetics
if 31<HbA1c<42:
  diabetic="normal"
elif 53>HbA1c>42:
  diabetic="pre-diabetic"
elif 75<HbA1c<100:
  diabetic=" Sunita you have diabetic"
else:
  diabetic="invalid"


#for kidney function
if 15>eGFR>0:
  kidney_function="kidney failure"
elif 60>eGFR>16:
  kidney_function="you have a kidney disease"
elif 90>eGFR>60:
  kidney_function="you have an early stage kidney disease"
elif 120>eGFR>90:
  kidney_function="you have Normal kidney"
else:
  kidney_function="invalid"

  
#for hemoglobin
if 11.6<hemoglobin<15:
  hemoglobin_level="you have a normal hemoglobin"
elif 0<hemoglobin<11.5:
  hemoglobin_level=" you have low hemoglobin"
else:
  hemoglobin_level="you have high hemoglobin"


# Print summary
print("\n\t" + initials)
print("\tweight = {}kg".format(weight))
print("\theight = {}m".format(height))
print("\tage = {} years old".format(age))
print("\tblood pressure = {}/{}".format(systolic, diastolic))
print("\tCholestrol = {} mg/dl".format(LDL,HDL))
print("\tthyroid  = {} mIU/L".format(TSH))
print("\tHbA1c  = {} mmol/mol".format(HbA1c))
print("\teGFR  = {} ".format(eGFR))
print("\themoglobin  = {} g/dl".format(hemoglobin))
print("\n\tbmi = {}".format(bmi))
print("\tpredicted maximal heart rate = {} bpm".format(max_HR))
print("\tblood pressure = " + bprisk)
print("\tcholestrol={} ".format(cholestrol))
print("\turic_acid={} ".format(uric_acid))
print("\tthyroid  = {} ".format(thyroid))
print("\tdiabetic  = {} ".format(diabetic))
print("\tkidney_function  = {} ".format(kidney_function))
print("\themoglobin  = {} ".format(hemoglobin_level))
print("\n")
```
